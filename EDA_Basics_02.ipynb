**Bivariate Analysis**
This analysis helps to identify how two features are correlated to each each other. i.e. whether Correlation between 2 feature is +ve or -ve. Correlatio is positive when on increasing value of one feature another also increases, and -ve when increasing value of one;another decrease. The correlation is 0 when the two features are independent of each other. 

Biavariate analysis show relation between features and target. IF two featureas hightly correlated like 95% same then we can drop one of the feature as they results in same result.

```python
sns.FacetGrid(df,hue='target',height=10).map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='target',height=10).map(plt.scatter,"petal_length","sepal_width").add_legend()
plt.show()
```
